```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
def append_value(dict_obj, key, value):
    # Check if key exist in dict or not
    if key in dict_obj:
        # Key exist in dict.
        # Check if type of value of key is list or not
        if not isinstance(dict_obj[key], list):
            # If type is not list then make it list
            dict_obj[key] = [dict_obj[key]]
        # Append the value in list
        dict_obj[key].append(value)
    else:
        # As key is not in dict,
        # so, add key-value pair
        dict_obj[key] = value
```

```python
org_tweets = {}

for index in tweets_df.index:
    sentencas = sent_tokenize(tweets_df.loc[index]['text'])
    palavras = word_tokenize(tweets_df.loc[index]['text'].lower())
    
    append_value(org_tweets, tweets_df.loc[index]['user_name'], palavras)

print(org_tweets)
```

```python
from nltk.corpus import stopwords
from string import punctuation

to_remove = []
to_remove = stopwords.words('portuguese')
to_remove.extend(list(punctuation))

stopwords = to_remove

for palavras in org_tweets:
    org_tweets[palavras] = [palavra for palavra in org_tweets[palavras] if palavra not in stopwords]
```
